```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import joblib
import os

os.makedirs('../MODELS', exist_ok=True)

# Load
events = pd.read_csv("../DATA/Astram event data_anonymized - Astram event data_anonymizedb40ac87.csv")

# Fill nulls
for col in ['event_type', 'event_cause', 'zone', 'junction', 'corridor']:
    events[col] = events[col].fillna('Unknown')
events['requires_road_closure'] = events['requires_road_closure'].fillna(False).astype(int)
events['priority'] = events['priority'].fillna('Low')

# Features
features = ['event_type', 'event_cause', 'zone', 'corridor']

# Label encode
le_dict = {}
X = events[features].copy()
for col in features:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    le_dict[col] = le

# Model 1 - Road Closure
y1 = events['requires_road_closure']
X_train, X_test, y_train, y_test = train_test_split(X, y1, test_size=0.2, random_state=42)
model_closure = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
model_closure.fit(X_train, y_train)
print("=== Road Closure Model ===")
print(classification_report(y_test, model_closure.predict(X_test)))

# Model 2 - Priority
y2 = (events['priority'] == 'High').astype(int)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y2, test_size=0.2, random_state=42)
model_priority = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
model_priority.fit(X_train2, y_train2)
print("=== Priority Model ===")
print(classification_report(y_test2, model_priority.predict(X_test2)))

# Impact Score - data derived
closure_rate = events.groupby('event_cause')['requires_road_closure'].mean()
high_priority_rate = events.groupby('event_cause').apply(lambda x: (x['priority']=='High').mean())
impact_scores = (0.6 * closure_rate + 0.4 * high_priority_rate) * 100
print("=== Impact Scores ===")
print(impact_scores.sort_values(ascending=False))

# Save
joblib.dump(model_closure, '../MODELS/model_closure.pkl')
joblib.dump(model_priority, '../MODELS/model_priority.pkl')
joblib.dump(le_dict, '../MODELS/le_dict.pkl')
joblib.dump(impact_scores, '../MODELS/impact_scores.pkl')
print("\nAll models saved!")
```
